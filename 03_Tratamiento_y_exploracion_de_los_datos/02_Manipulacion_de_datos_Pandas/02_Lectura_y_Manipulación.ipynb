# Manipulación y lectura de datos en pandas

## ¿Para qué sirve?

La carga y lectura de datos en Pandas sirve para importar y manipular datos dacilmente dentro de Python. Es uno de los pasos más importantes cuando trabajas con analisis de datos, ciencia de datos o cualquier proceso relacionado con bases de datos y archivos.

Asi se importa la libreria de pandas y se hace la lectura de cualquier archivo mencionado en la diapositiva.

La carga de lectura de datos en Pandas sirve para importar y manipular datos fácilmente...
