## A√±adir dos nuevas columnas

No te quedes solo con los datos que te dan. En su lugar, se pueden a√±adir nuevas columnas aun DataFrame. SA esto se le llama `Transformaci√≥n`, `Mutaci√≥n` e `Ingenier√≠a de Caracteristicas`.

Se pueden crear columnas nuevas desde cero, pues es habitula derivarlas de otras columnas, por ejemplo, sumando columnas o cambiando sus unidades.

**Instrucciones**

- A√±ade una nueva columna a homelessness, denominada `total`, que contenga la suma de las columnas individuales y family_members
- A√±ade otra columna a homelessness, denominada p_homeless, que contenga la proporci√≥n de la poblaci√≥n de personas sin hogar total con respecto a la poblaci√≥n total de cada estado stage_pop.

```python
# Agregar una columna total como la suma de individuals y family_members
homelessness['total'] = homelessness['individuals'] + homelessness['family_members']

# Agregar una columna p_homeless como proporci√≥n del total de homeless
# de la poblaci√≥n para el estado de la poblaci√≥n (state_pop)

homelessness['p_homeless'] = homelessness['total'] / homelessness['state_pop']

# Visualizar el resultado
print(homelessness.head(10))
```

En este ejercisio, responderas la pregunta: `"¬øQu√© estado tiene el mayor numero de personas in hogar por cada 10,000 habitantes"`. Convina las nuevas habilidades en pandas para averiguarlo

**Intrucciones**

- A√±ade una columna a homelessness, indiv_per_10k, que contenga el numero de personas sin hogar por cada 10,000 de personas en cada estado, utilizando state_pop para la poblaci√≥n del estado
- Subconjunta las filas en las que el indiv_per_10k sea superiora 20 asignandolas a high_homelessness
- Ordena high_homelessness en orden desendente indiv_per_10k, asigando a high_homelessness_srt. Seleccionando solo las columnas state e indiv_per_10k de high_homelessness_srt y guardar como result y Visualiza result

```python
# Crear la columna indiv_per_10k

homelessness['indiv_per_10k'] =  10000 * homelessness ['individuals'] / homelessness['state_pop']

# Subconjuntar filas para indiv_per_10k mayores que 20

high_homelessness = homelessness[homelessness['indiv_per_10k']>20]

# Ordenar hight_homelessness por u¬øindiv_per_10k
result = high_homelessness.sort_values('indiv_per_10k', ascending=False)

# Visualizar el resultado
result
```
